```python
import pandas as pd
```

```python
data={'Title':[],'price':[],'stock':[],'pic_url':[]}
for i in range(10):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    website = requests.get(url).text
    website = BeautifulSoup(website , "lxml")
    books = website.find_all("li", class_ = "col-xs-6 col-sm-4 col-md-3 col-lg-3")
    for bok in books: 
        price = bok.find("p", class_ = "price_color").text
        title = bok.find("h3").find("a")["title"]
        stock = bok.find("p", class_ = "instock availability").text
        clean_stock = stock.strip()
        clean_stock = clean_stock.replace('\n', '').strip()
        pic_url = bok.find("a")
        pic=pic_url.find('img')
        pic_src=pic['src']
        data["Title"].append(title)
        data["price"].append(price)
        data["stock"].append(clean_stock)
        data["pic_url"].append(pic_src)
df=pd.DataFrame(data)
df
```
